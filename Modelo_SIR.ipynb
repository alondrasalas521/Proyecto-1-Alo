```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Stl(St, It, aplha):
  return St - alpha * St * It

def Itl(It, St, aplha, gamma):
  return It + alpha * It * St - gamma * It

def Rtl(Rt, It, gamma):
  return Rt + gamma * It
```

```python
#condiciones iniciales

S0 = 100
I0 = 1
R0 = 0

alpha = 0.005
gamma = 0.1

T = np.linspace(0, 99, 100)

S = [S0]
I = [I0]
R = [R0]


for t in T :
  S.append(Stl(S[-1], I[-1], alpha))
  I.append(Itl(I[-1], S[-1], alpha, gamma))
  R.append(Rtl(R[-1], I[-1], gamma))

plt.plot( T, S[ : -1], label = "Suceptibles")
plt.plot( T, I[ : -1], label = "Infectados")
plt.plot( T, R[ : -1], label = "Recuperados" )
plt.legend()
plt.show()
```
